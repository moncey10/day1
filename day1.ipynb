```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel("data.xlsx")

df = df.iloc[:,0].str.split(",", expand=True)

df.columns = ["user_id","skill_id","skill_title","rating","rating_date","department","location"]

df["rating"] = pd.to_numeric(df["rating"], errors='coerce')
df["rating_date"] = pd.to_datetime(df["rating_date"], errors='coerce')


df.drop_duplicates(inplace=True)

df.dropna(inplace=True)

print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
print("Duplicates:", df.duplicated().sum())

plt.figure(figsize=(8,5))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.show()

df.to_excel("cleaned_data.xlsx", index=False)
```

Summary of Data Issues and Cleaning Actions:

- Duplicates: None found.
- Missing Values: None found.
- Column types: All correct (rating  int, rating_date  datetime).
- Result: Cleaned DataFrame ready for analysis.
